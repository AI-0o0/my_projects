### ğŸ“Š Data Types in NumPy

| Code | Data Type              | Description                                 |
|------|------------------------|---------------------------------------------|
| `i`  | Integer                | Signed integer types                        |
| `b`  | Boolean                | Boolean values (`True` or `False`)          |
| `u`  | Unsigned Integer       | Non-negative integers                       |
| `f`  | Float                  | Floating-point (decimal) numbers            |
| `c`  | Complex Float          | Complex numbers with float real/imag parts  |
| `m`  | Timedelta              | Differences between dates/times             |
| `M`  | Datetime               | Date and time values                        |
| `O`  | Object                 | Python objects                              |
| `S`  | String (byte)          | Fixed-length ASCII byte strings             |
| `U`  | Unicode String         | Fixed-length Unicode strings                |
| `V`  | Void                   | Raw data (fixed-size memory chunk)          |


```python
arr = np.array([1, 2, 3, 4])
print(arr)
print(arr.dtype)

arr = np.array(arr, dtype='S')
print(arr)
print(arr.dtype)

#For i, u, f, S and U we can define size as well.
arr = np.array(arr, dtype='i2')
print(arr)
print(arr.dtype)

#arr = np.array(['a', '2', '3'], dtype='i') this produse an error

arr = np.array([1.1, 2.1, 3.1])
print(arr.dtype)

newarr = arr.astype('i')
print(newarr)
print(newarr.dtype)

arr = np.array([-1, 0, 1])
newarr = arr.astype(bool)
print(newarr)
```

```python
#copy
arr = np.array([1, 2, 3, 4, 5])
x = arr.copy()
arr[0] = 42

print(arr)
print(x)
print(x.base)

#view
arr = np.array([1, 2, 3, 4, 5])
x = arr.view()
arr[0] = 42
x[1]=404

print(arr)
print(x)
print(x.base)
```

```python
arr = np.array([41, 42, 43, 44])

# Create an empty list
filter_arr = []

# go through each element in arr
for element in arr:
  # if the element is higher than 42, set the value to True, otherwise False:
  if element > 42:
    filter_arr.append(True)
  else:
    filter_arr.append(False)

newarr = arr[filter_arr]

print(filter_arr)
print(newarr)
```

```python
arr = np.array([1, 2, 3, 4, 5, 6, 7])

# Create an empty list
filter_arr = []

# go through each element in arr
for element in arr:
  # if the element is completely divisble by 2, set the value to True, otherwise False
  if element % 2 == 0:
    filter_arr.append(True)
  else:
    filter_arr.append(False)

newarr = arr[filter_arr]

print(filter_arr)
print(newarr)
```

```python
x = [1, 2, 3, 4]
y = [4, 5, 6, 7]
z = np.add(x, y)
print(z)

def myadd(x, y,z):
  return x+y+z
myadd = np.frompyfunc(myadd, 3, 1)
print(myadd([1, 2, 3, 4], [5, 6, 7, 8], [1,1,1,1]))

arr1 = np.array([10, 20, 30, 40, 50, 60])
arr2 = np.array([3, 21, 22, 23, 24, 25])
newarr = np.subtract(arr1, arr2)
print(newarr)

newarr = np.multiply(arr1, arr2)
print(newarr)

newarr = np.divide(arr1, arr2)
print(newarr)

newarr = np.power(arr1, arr2)
print(newarr)

newarr = np.mod(arr1, arr2)
print(newarr)

newarr = np.remainder(arr1, arr2)
print(newarr)

newarr = np.divmod(arr1, arr2)
print(newarr)

arr = np.array([-1, -2, 1, 2, 3, -4])
newarr = np.absolute(arr)
print(newarr)

arr = np.trunc([-3.1666, 3.6667])
print(arr)

arr = np.fix([-3.1666, 3.6667])
print(arr)

arr = np.around(3.1666, 2)
print(arr)
```
